```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    for likes in likes_list:
        total_likes += likes
    
    print("Total Likes:", total_likes)
    
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example
analyze_post_engagement([200, 300, 250, 300])
```

```python
def medicine_stock_checker(stock):
    print("Medicine Stock:", stock)
    
    if stock < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")


# Example
medicine_stock_checker(6)
```

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    for rain in rainfall_data:
        total += rain
    
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", average)
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example
rainfall_checker([70, 75, 68, 75], 70)
```

```python
def duplicate_account_checker(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example
duplicate_account_checker(["user1", "user2", "user1"])
```

```python
def premium_crop_filter(prices):
    premium = []
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)


# Example
premium_crop_filter([1500, 2500, 1800, 3200])
```

```python
def system_health_checker(error_count):
    print("Error Count:", error_count)
    
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example
system_health_checker(7)
```

```python
def attendance_checker(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example
attendance_checker(["P", "P", "A", "P", "P"])
```

```python
def electricity_bill_calculator(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status


# Example
bill, usage_status = electricity_bill_calculator(250)
print("Total Bill:", bill)
print("Usage Level:", usage_status)
```
