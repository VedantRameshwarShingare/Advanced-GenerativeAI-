```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        amount = int(input("Enter recharge amount: ₹"))
        
        if amount < 50:
            print("Recharge amount must be at least ₹50.")
            continue
        
        if amount not in valid_plans:
            print("Invalid plan selected. Valid plans are:", valid_plans)
            continue
        
        print("Recharge successful!")
        break

validate_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Eg
inventory = {
    "Laptop": 20,
    "Mouse": 10,
    "Keyboard": 25,
    "Monitor": 12
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0
    
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Eg
marks_list = [60, 55, 70, 45]
result = process_result(marks_list)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak_hour:
        fare += fare * 0.25  # 25% extra
    
    return fare

while True:
    try:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        
        peak_hour = True if peak == "yes" else False
        
        total_fare = calculate_fare(km, peak_hour)
        print("Total Fare: ₹", total_fare)
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
    
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
```

```python
# Problem 5: Employee Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / len(attendance_list)) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Eg
attendance = ["P", "P", "A", "P", "P", "P", "P", "P"]
status = check_attendance(attendance)
print("Eligibility:", status)
```

```python
# Problem 6: Password Strength Checker

def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    has_digit = False
    has_special = False
    
    special_chars = "@#$"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character @#$)"

# Eg
pwd = input("Enter password: ")
print(check_password_strength(pwd))
```
