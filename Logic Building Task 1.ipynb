```python
#Logic Building Task 1


# 1️ User Login Check
correct_username = "admin"
correct_password = "1234"

# User input 
username = "admin"
password = "1234"


if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python

# 2️ Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
# 3️ Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()  
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)

```

```python

# 4️ Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")
    
    if length > 10:
        print("⚠️  This message is longer than 10 characters")


```

```python
#️ 5 Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
